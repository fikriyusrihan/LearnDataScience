**Climb The Ladder** 
<br>
<br>
Sebuah bangunan memiliki beberapa tingkat, kita akan mencoba menghitung sampai pada lantai berapakah kita dalam 100x percobaan pelembaran dadu dengan ketentukan sebagai berikut:


1.   Jika dadu bernilai ganjil, maka kita akan turun satu lantai.
2.   Jika dadu bernilai genap, maka kita naik satu lantai.
3.   Jika mendapat dadu bernilai 6, maka kita akan melempar dadu kembali dan naik sebanyak angka yang dikeluarkan dadu tersebut.
4.   Jika berada dilantai 1, maka tidak akan turun.

Hitung berapa kemungkinan mencapai lantai 40, dengan 100x pelemparan dadu setelah 500 percobaan.



```
# import seluruh library yang dibutuhkan
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# mengatur nilai seed
np.random.seed(123)

# mencatat setiap perubahan tingkatan lantai
tot_step = []

# melalukan perulangan
for i in range(500):
  step = [1]
  for j in range(100):
    dice = np.random.randint(1, 7)

    if dice % 2 == 0: # dadu bernilai genap
      if dice == 6:
        step.append(step[-1] + np.random.randint(1,7))
      else:
        step.append(step[-1] + 1)
    else:
      step.append(max(1, step[-1] - 1))

  tot_step.append(step)

np_tot_step = np.array(tot_step)
np_tot_step_tp = np.transpose(np_tot_step)

# melakukan seleksi pada data
greater_equal_40 = np_tot_step[np_tot_step[:, -1] >= 40]

# melakukan visualisasi
plt.plot(np.transpose(greater_equal_40))
plt.show()

# menghitung probabilitas
print("Peluang: {}%".format((greater_equal_40.size / 101) / 500 * 100))
```
